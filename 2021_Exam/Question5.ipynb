```python
import numpy as np
import matplotlib.pyplot as plt
import math
lambda0 = 1
lambda1 = 2

i = np.arange(0, 10, 0.1)
#When I1 != I2
#fz=λX*λY/(λY−λX )*(exp(−λX*i)−exp(−λY*i))

#I1
fx = np.exp(-i * lambda0) * lambda0
#I2
fy = np.exp(-i * lambda1) * lambda1
#I1+I2
fz = (lambda0 * lambda1)/(lambda1 - lambda0)*(np.exp(-i * lambda0)-np.exp(-i * lambda1))

plt.figure(figsize=(10, 8))
plt.plot(i, fx, label="I1")
plt.plot(i, fy, label="I2")
plt.plot(i, fz, label="I1+I2")

plt.legend()
plt.show()
```
