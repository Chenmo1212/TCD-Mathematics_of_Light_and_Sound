```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()
#Generate 100000 random vectors in the range -5, 5
x=np.zeros(100000)
y=np.zeros(100000)
for i in range(0,100000):
    x[i] = np.random.uniform(-5,5.0)
    y[i] = np.random.uniform(-5,5.0)
X=np.zeros(10000)
Y=np.zeros(10000)
#calcutale the sum of 10 vectors
for i in range(0,10000):
    for j in range(0,10):
        X[i] += x[10 * i + j]
        Y[i] += y[10 * i + j]
```

```python
#Make the unit length of the x,y axis equal
ax = plt.gca()
ax.set_aspect(1)
#Set the axis range
plt.xlim(-50,50)
plt.ylim(-50,50)
plt.scatter(X,Y,s=0.1,c='red')

plt.show()
```
